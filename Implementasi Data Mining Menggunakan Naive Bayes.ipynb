```python
#Memanggil library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca Data Kelulusan Mahasiswa
mhs = pd.read_excel("Data Kelulusan Mahasiswa.xlsx")
mhs.head(10)
```

```python
x = mhs.drop(["StatusKelulusan"], axis=1)
x.head(10)
```

```python
y = mhs["StatusKelulusan"]
y.head(10)
```

```python
# Import Gaussian Naive Bayes Model
from sklearn.naive_bayes import GaussianNB

# Memanggil fungsi Naive Bayes
nbc = GaussianNB()

# Memasukkan data training pada fungsi klasifikasi naive bayes
data_training = nbc.fit(x,y)

# Melakukan prediksi pada data training
y_predict = data_training.predict(x)
print(y_predict)
```

```python
# Melakukan Prediksi pada Data Input => Data yang akan diprediksi
Jenis_Kelamin = 1 #Laki-Laki
Status_Kerja = 0 #Mahasiswa
Umur = 25
Status_Nikah = 1 #Menikah
IPS1 = 3.12
IPS2 = 2.75
IPS3 = 2.05
IPS4 = 3.01
IPS5 = 2.56
IPS6 = 3.67
IPS7 = 2.75
IPS8 = 3.45
IPK = np.average([IPS1,IPS2,IPS3,IPS4,IPS5,IPS6,IPS7,IPS8])
Data_Testing = [[Jenis_Kelamin, Status_Kerja, Umur, Status_Nikah, IPS1, IPS2, IPS3, IPS4, IPS5, IPS6, IPS7, IPS8, IPK]]

```

```python
# Melakukan prediksi berdasarkan Data Input
y_pred = data_training.predict(Data_Testing)
```

```python
# Mencetak Hasil Prediksi Kelulusan Mahasiswa
# Hasil 0 adalah Terlambat
# Hasil 1 adalah Tepat

if y_pred == 0:
    hasil = "Terlambat"
elif y_pred == 1:
    hasil = "Tepat"
else:
    hasil == "Error"
print("Hasil Prediksi Kelulusan Mahasiswa :", hasil)
```

# Menghitung Akurasi => melihat skor akurasi
from sklearn.metrics import accuracy_score
print("Akurasi = %0.2f" % accuracy_score(y, y_predict))

```python
# Mencetak Laporan Hasil Klasifikasi
from sklearn.metrics import classification_report
print(classification_report(y, y_predict))
```
